# Loading Data

Data we want to analyze will usually come in an external file on a tabular form.
To load this type of data in our notebook we need to use a library. 
A library is an external collection of different data types and operators that someone built for an specific purpose.
Python has many libraries, the one used for working with data is called Pandas

To be able to use a library, first of all we need to import the library to our notebook. 

once the library is imported we have access to all the data structures and functions that it includes

```python
import pandas as pd
```

```python
orders = pd.read_csv('DataSources/01-OrderData.csv')
```

```python
orders
```

### And this is Excel in steroids
#### Here we can
* Reference data
* Filter data
* Aggregate data 
All with a single command. We can enve perfom the three operations at a time with a single command

```python
orders.head()
```

```python
orders.tail()
```

lets say we want to describe the data by the STATUS value

```python
orders.groupby('STATUS').count()
```

Can we do better?

```python
orders[orders['STATUS']=='Cancelled'].groupby(['ClientID','PRODUCTLINE']).agg({'SALES':['count','sum','max','min']})
```

```python
orders['Date'] = pd.to_datetime(orders['Date'])
```

```python
orders[orders['STATUS']=='Shipped'].groupby([orders['Date'].dt.year,'PRODUCTLINE']).agg({'SALES':['count','sum','max','min']})
```

## Using Tabula

We can use external lybraries to do some complex tasks such as incorporating data that is in PDF format and get our result

```python
import tabula as tab
```

```python
AddressBook = tab.read_pdf("DataSources/02-ClientsDirectory.pdf", pages='all')
```

```python
len(AddressBook)
```

```python
mergeAddress = pd.DataFrame()
```

```python
for table in AddressBook:
    mergeAddress = mergeAddress.append(table)
    
```

```python
mergeAddress
```

```python
orders = pd.merge(orders,mergeAddress[['ClientID','ClientName','Country','Region']],on="ClientID")
```
